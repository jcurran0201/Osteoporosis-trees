# Decision Tree Classifier

```python
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, accuracy_score
from sklearn.tree import DecisionTreeClassifier  
from sklearn import metrics 
from sklearn.metrics import (
    f1_score,
    accuracy_score,
    recall_score,
    precision_score,
    confusion_matrix,
    classification_report,
    roc_auc_score,
    precision_recall_curve,
    roc_curve,
    make_scorer,
)
```

```python
d_tree=DecisionTreeClassifier()
```

```python
X = new_df.drop(["Osteoporosis"], axis=1)
Y =  new_df['Osteoporosis'] # Complete the code to define the dependent (target) variable



# Splitting the data in 70:30 ratio for train to test data
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.20, random_state=1
)
```

```python
def metrics_score(actual, predicted):
    print(classification_report(actual, predicted))
    
    cm = confusion_matrix(actual, predicted)
    
    plt.figure(figsize = (8, 5))
    
    sns.heatmap(cm, annot = True,  fmt = '.2f', xticklabels = ['No Osteoporosis', 'Osteoporosis'], yticklabels = ['No Osteoporosis', 'Osteoporosis'])
    
    plt.ylabel('Actual')
    
    plt.xlabel('Predicted')
    
    plt.show()
```

```python
d_tree.fit(X_train,y_train)
```

```python
y_pred1 = d_tree.predict(X_train)
```

```python
metrics_score(y_train, y_pred1)
```

```python
y_pred_test1 = d_tree.predict(X_test)

metrics_score(y_test,y_pred_test1)
```

## Tune hyperparamaters in order to fix overfitting

```python
tuned_d_tree = DecisionTreeClassifier(random_state=20)
```

```python
#Set paramaters for the GridSearchCV()
params = {'max_depth': np.arange(2, 10), 
              'criterion': ['gini', 'entropy'],
              'min_samples_leaf': [5, 10, 20, 25]
             }
```

```python
scorer = metrics.make_scorer(recall_score,pos_label=1) 
grid_obj= GridSearchCV(tuned_d_tree,param_grid=params,scoring=scorer,cv=5)
```

```python
grid_obj = grid_obj.fit(X_train,y_train)
```

```python
tuned_d_tree = grid_obj.best_estimator_
```

```python
tuned_d_tree.fit(X_train,y_train)
```

```python
y_pred_train2 = tuned_d_tree.predict(X_train)

metrics_score(y_train,y_pred_train2)
```

```python
# Checking performance on the testing data
y_pred_test2 = tuned_d_tree.predict(X_test)

metrics_score(y_test,y_pred_test2)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_forest = RandomForestClassifier(criterion='entropy')
```

```python
X = new_df.drop(["Osteoporosis"], axis=1)
Y =  new_df['Osteoporosis'] 



# Splitting the data in 70:30 ratio for train to test data
X_train, X_test, y_train, y_test = train_test_split(
    X, Y, test_size=0.20, random_state=1
)
```

```python
random_forest.fit(X_train,y_train)
```

```python
y_pred2 = random_forest.predict(X_train)
```

```python
metrics_score(y_train,y_pred2)
```

```python
#Tune Hyperparams
```

```python
random_forest.fit(X_test,y_test)
```

```python
rf_estimator_tuned = RandomForestClassifier(criterion = "entropy", random_state = 7)
parameters = {"n_estimators": [110, 120],
    "max_depth": [6, 7],
    "min_samples_leaf": [20, 25],
    "max_features": [0.8, 0.9],
    "max_samples": [0.9, 1],
    "class_weight": ["balanced",{0: 0.3, 1: 0.7}]} 
scorer = metrics.make_scorer(recall_score, pos_label = 1)

# Run the grid search
grid_obj = GridSearchCV(rf_estimator_tuned, parameters, scoring = scorer, cv = 5)

grid_obj = grid_obj.fit(X_train, y_train)

# Set the classifier to the best combination of parameters
rf_estimator_tuned = grid_obj.best_estimator_
```

```python
rf_estimator_tuned.fit(X_test,y_test)
```

```python
y_pred_train4 = rf_estimator_tuned.predict(X_train)
metrics_score(y_train, y_pred_train4)
```

```python
y_pred_test4 = rf_estimator_tuned.predict(X_test)

metrics_score(y_test, y_pred_test4)
```

```python
importances = rf_estimator_tuned.feature_importances_

indices = np.argsort(importances)

feature_names = list(X.columns)

plt.figure(figsize = (12, 12))

plt.title('Feature Importances')

plt.barh(range(len(indices)), importances[indices], color = 'violet', align = 'center')

plt.yticks(range(len(indices)), [feature_names[i] for i in indices])

plt.xlabel('Relative Importance')

plt.show()
```
